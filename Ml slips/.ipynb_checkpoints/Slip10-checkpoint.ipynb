```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target labels
target_names = iris.target_names
```

```python
# Step 1: Apply PCA to reduce dimensions to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Create a DataFrame with PCA results and labels for easy plotting
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['Target'] = y
```

```python
# Step 2: Plot the data in 2D using the principal components
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_pca, x='PC1', y='PC2', hue='Target', palette='viridis', s=100, alpha=0.7
)
plt.title("PCA of Iris Dataset (2 Components)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(target_names, title="Target")
plt.show()
```

```python
# Step 3: Display the explained variance ratio for each component
print("Explained Variance Ratio of Each Component:")
print(pca.explained_variance_ratio_)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Add the target variable and convert categorical values to numeric
iris_df['species'] = iris.target
```

```python
# Map numeric values back to the species names
species_map = dict(enumerate(iris.target_names))
iris_df['species_name'] = iris_df['species'].map(species_map)
```

```python
# Print the dataset with numeric and categorical species names
print(iris_df.head())
```

```python
# Scatter plot using seaborn
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species_name', data=iris_df, palette='viridis')
plt.title('Scatter Plot of Sepal Length vs Sepal Width (Iris Dataset)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```
